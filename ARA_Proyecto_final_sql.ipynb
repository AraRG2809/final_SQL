#**12**
¿Cuál es el director que hizo más películas?

```python

cur.execute('''
    SELECT d.name, COUNT(m.id) AS movie_count
    FROM directors d
    JOIN movies m ON d.id = m.director_id
    GROUP BY d.name
    ORDER BY movie_count DESC
    LIMIT 1
''')
movies = cur.fetchall()
movies = pd.DataFrame(movies, columns = ['name', 'movie_count'])
movies.head(3)
```

```python
#@title Comprueba el ejercicio
def check():
  if movies.iloc[0,0]== 'Steven Spielberg':
    return 'Correcto'
  else:
    return 'Incorrecto'
check()
```
